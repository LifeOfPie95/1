```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
s1 = pd.read_csv('sensor_10.csv',sep = ';')
s2 = pd.read_csv('sensor_13.csv',sep = ';')
s3 = pd.read_csv('sensor_15.csv',sep = ';')
s4 = pd.read_csv('sensor_16.csv',sep = ';')
s5 = pd.read_csv('sensor_17.csv',sep = ';')
s6 = pd.read_csv('sensor_18.csv',sep = ';')
times1 = pd.DatetimeIndex(s1['Date & Time'])
times2 = pd.DatetimeIndex(s2['Date & Time'])
times3 = pd.DatetimeIndex(s3['Date & Time'])
times4 = pd.DatetimeIndex(s4['Date & Time'])
times5 = pd.DatetimeIndex(s5['Date & Time'])
times6 = pd.DatetimeIndex(s6['Date & Time'])
pm2p5s1 = s1['PM2.5']
pm2p5s2 = s2['PM2.5']
pm2p5s3 = s3['PM2.5']
pm2p5s4 = s4['PM2.5']
pm2p5s5 = s5['PM2.5']
pm2p5s6 = s6['PM2.5']

date_1 = '01/01/2022'

sub_s1 = pm2p5s1[times1 > date_1]
sub_t1 = times1[times1 > date_1]

sub_s2 = pm2p5s2[times2 > date_1]
sub_t2 = times2[times2 > date_1]

sub_s3 = pm2p5s3[times3 > date_1]
sub_t3 = times3[times3 > date_1]


plt.plot(sub_t1,sub_s1)
plt.plot(sub_t2,sub_s2)
plt.plot(sub_t3,sub_s3)

av_s1 = np.nanmean(sub_s1)
av_s2 = np.nanmean(sub_s2)
av_s3 = np.nanmean(sub_s3)

std_s1 = np.nanstd(sub_s1)
std_s2 = np.nanstd(sub_s2)
std_s3 = np.nanstd(sub_s3)

print(av_s1,av_s2)
print(std_s1,std_s2)


```
